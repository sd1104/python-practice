```python
def fn5(x):
    return np.sqrt(x)

plt.xlim(0, 50)
plt.ylim(0,10)
x = np.linspace(0,50)
y = fn5(x)

plt.plot(x,y)
plt.show()
```

```python
def fn5(x):
    return np.sqrt(x)+1

plt.xlim(0, 50)
plt.ylim(0,10)
x = np.linspace(0,50)
y = fn5(x)

plt.plot(x,y)
plt.show()
```

```python
def fn6(x):
    return 3*x**2 + 2*x + 1

plt.xlim(-5, 5)
plt.ylim(0,50)
x = np.linspace(-5,5)
y = fn6(x)

plt.plot(x,y)
plt.show()
```

```python
def fn7(x):
    return 4*x**3+ 2*x**2 + x + 3

plt.xlim(-5, 5)
plt.ylim(-50,50)
x = np.linspace(-5,5)
y = fn7(x)

plt.plot(x,y)
plt.show()
```

```python
def fn8(x):
    return np.sin(x)

def fn9(x):
    return np.cos(x)


plt.xlim(-4, 4)
plt.ylim(-1.5,1.5)
x = np.linspace(-np.pi, np.pi)
y1 = fn8(x)
y2 = fn9(x)

plt.plot(x,y1,label='sin')
plt.plot(x,y2,label='cos')
plt.legend()
plt.show()
    
```

```python
def fn10(x):
    return np.tan(x)


x = np.linspace(-1.3, 1.3)
y3 = fn10(x)
plt.plot(x,y3,label='tan')
plt.legend()
plt.show()
```
